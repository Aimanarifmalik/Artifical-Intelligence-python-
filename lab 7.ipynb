```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
file = "titanic.csv"
df = pd.read_csv(file)

# Convert categorical features to numeric format
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

# Print column names
print(df.columns)

# Define features and target variable
features = ['Pclass', 'Sex_male', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_Q', 'Embarked_S']
X = df[features]
y = df['Survived']

# Split dataset into training and testing sets using a specific random state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the KNN model
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
    
# Evaluate model's accuracy on testing set
accuracy = knn.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```
