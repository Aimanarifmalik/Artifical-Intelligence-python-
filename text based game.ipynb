```python
import time

inventory = []

def add_to_inventory(item):
    inventory.append(item)
    print("You added", item, "to your inventory.")

def remove_from_inventory(item):
    if item in inventory:
        inventory.remove(item)
        print("You removed", item, "from your inventory.")
    else:
        print("You don't have", item, "in your inventory.")

def show_inventory():
    print("Your inventory:", inventory)

def start_game():
    print("You wake up in the heart of an enchanted grove.")
    time.sleep(2)
    print("You must find a way out of this mysterious forest.")
    time.sleep(2)
    print("You see a path leading deeper into the forest.")

start_game()

while True:
    action = input("What will you do? (Type 'help' for options): ").lower()

    if action == "help":
        print("Commands: 'go', 'add', 'remove', 'inventory', 'quit'")

    elif action == "go":
        direction = input("Which path will you take? (Sunlit Glade or Shadowy Thicket): ").lower()
        if direction == "sunlit glade":
            print("You walk towards the Sunlit Glade...")
            time.sleep(2)
            print("You reach the Sunlit Glade.")
        elif direction == "shadowy thicket":
            print("You venture into the Shadowy Thicket...")
            time.sleep(2)
            print("You reach the Shadowy Thicket.")
        else:
            print("Invalid path. Choose again.")

    elif action == "add":
        item = input("What item do you want to add to your inventory? ").lower()
        add_to_inventory(item)

    elif action == "remove":
        item = input("What item do you want to remove from your inventory? ").lower()
        remove_from_inventory(item)

    elif action == "inventory":
        show_inventory()

    elif action == "quit":
        print("Thanks for playing!")
        break

    else:
        print("Invalid command. Type 'help' for options.")
```
