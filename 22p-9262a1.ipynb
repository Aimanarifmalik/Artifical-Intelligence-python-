```python
import os
import math

directory = r'C:\Users\Aiman\Desktop\P22-9262'

file_name = 'input1.txt'

file_path = os.path.join(directory, file_name)

with open(file_path, 'r') as file:
    text = file.read()

#initializing variables for counting

numof_chars = 0


numof_words = 0

numof_sentences = 0

terminators = ['.', ':', ';', '?', '!']


for char in text:
    if char.isalnum():
        numof_chars = numof_chars + 1
    elif char in terminators:
        numof_sentences = numof_sentences + 1

# for the case in which file has no terminators and only one sentence
if numof_sentences == 0:
    numof_sentences = 1

# Split the text into words and count them
words = text.split()
numof_words = len(words)


print("\n\nNumber of characters :\t", numof_chars)

print("\n\nNumber of words :\t", numof_words)

print("\n\nNumber of sentences : \t", numof_sentences)



index = 4.71*(numof_chars / numof_words) + 0.5*(numof_words/numof_sentences) - 21.43

#rounding of to the nearest integer

rounded_index = math.ceil(index)

print("\n\nReadability Index : \t", rounded_index)

print("\n\n")
```
